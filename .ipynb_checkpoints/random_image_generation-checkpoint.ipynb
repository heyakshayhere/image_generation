```python
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import random
import warnings
warnings.filterwarnings("ignore")
```

```python
fig = plt.figure(figsize=(20, 20))  #height*width

img_size = (225, 225)      #height*width

columns = rows = 5

try:
    for i in tqdm(range(1, columns*rows +1)):
        img = get_random_image(img_size)
        fig.add_subplot(rows, columns, i)
        plt.axis(False)
        plt.imshow(img,interpolation='nearest',resample=random.randint(1,1000),interpolation_stage = 'rgba')
    plt.savefig('all_plots.jpg',vmax='I', vmin='I', hmax='I', hmin='I',bbox_inches='tight',quality=100, optimize=True,progressive=True,pad_inches=.2)
    plt.show()
except:
    pass
```
